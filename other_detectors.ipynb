# Run `delly`

## Installing `bcftools`

To install bcftools, run the following commands:
```bash
sudo apt-get update  
sudo apt-get install -y bcftools
```

## Installing and running `delly`

````python
%%bash

./configure_delly.sh
````

````python
%%bash

mkdir -p output
./delly_v0.8.7_linux_x86_64bit call -o output/delly.bcf -g data/GRCh38_full_analysis_set_plus_decoy_hla.fa data/SRR_final_sorted.bam
bcftools view output/delly.bcf > output/delly.vcf
````

# Run `read_depth`

````python
%%bash

./configure_CNVpytor.sh
````

````python
%%bash

cnvpytor -root file.pytor -rd data/SRR_final_sorted.bam -chrom $(seq 1 22) X Y
cnvpytor -root file.pytor -fasta data/GRCh38_full_analysis_set_plus_decoy_hla.fa -his 1000 10000 100000
cnvpytor -root file.pytor -stat 1000 10000 100000
cnvpytor -root file.pytor -call 1000 10000 100000
cnvpytor -root file.pytor -view 1000 
set print_filename output/cnvpytor_1000.vcf
print calls
cnvpytor -root file.pytor -view 10000 
set print_filename output/cnvpytor_10000.vcf
print calls
cnvpytor -root file.pytor -view 100000 
set print_filename output/cnvpytor_100000.vcf
print calls
````
